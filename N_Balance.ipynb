```python
import pandas as pd
import matplotlib.pyplot as plt

col_time          = "Time and Date"
col_n_input       = "output.Balance.Soil N Balance.N Input[kg N/ha]"
col_n_mineralized = "output.Balance.Soil N Balance.Nmin mineralized [kg N/ha]"
col_n_immobilized = "output.Balance.Soil N Balance.Nmin immobilized [kg N/ha]"
col_n_uptake      = "output.Balance.Soil N Balance.Nmin Uptake [kg N/ha]"
col_gaseous       = "output.Balance.Soil N Balance.Gaseous N Emission [kg N/ha]"
col_leaching      = "output.Balance.Soil N Balance.N Leaching [kg N/ha]"

# --- Convert to cumulative values (if daily values, cumsum makes them cumulative) ---
df["N_input_cum"]       = df[col_n_input].cumsum()
df["N_mineralized_cum"] = df[col_n_mineralized].cumsum()
df["N_immobilized_cum"] = df[col_n_immobilized].cumsum()
df["N_uptake_cum"]      = df[col_n_uptake].cumsum()
df["N_gaseous_cum"]     = df[col_gaseous].cumsum()
df["N_leaching_cum"]    = df[col_leaching].cumsum()

# --- Nitrogen balance equations ---
df["DeltaN_simple"] = df["N_input_cum"] - (
    df["N_uptake_cum"] + df["N_gaseous_cum"] + df["N_leaching_cum"]
)

df["DeltaN_detailed"] = (
    df["N_input_cum"] + df["N_mineralized_cum"]
    - (df["N_uptake_cum"] + df["N_immobilized_cum"]
       + df["N_gaseous_cum"] + df["N_leaching_cum"])
)

# --- Print final cumulative values ---
print("Final cumulative values:")
print(df[["N_input_cum","N_mineralized_cum","N_immobilized_cum",
          "N_uptake_cum","N_gaseous_cum","N_leaching_cum",
          "DeltaN_simple","DeltaN_detailed"]].iloc[-1])

# --- Plot ---
plt.figure(figsize=(10,6))
plt.plot(df[col_time], df["N_input_cum"], label="N Input (cum)")
plt.plot(df[col_time], df["N_uptake_cum"], label="N Uptake (cum)")
plt.plot(df[col_time], df["N_gaseous_cum"], label="Gaseous N (cum)")
plt.plot(df[col_time], df["N_leaching_cum"], label="Leaching (cum)")
plt.plot(df[col_time], df["DeltaN_detailed"], label="DeltaN (detailed)", linewidth=2)
plt.xlabel("Time")
plt.ylabel("kg N/ha")
plt.title("Nitrogen Balance")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- your two simulation files ---
file1 = r"C:\Users\moham\OneDrive\Desktop\water in Soil Plant System\Exercise\Exercise 8\completed\EC2_N-Balance\EC2_N-Balance\output\EC2_N-Balance_1_1_1_0_MIneral.xno"
file2 = r"C:\Users\moham\OneDrive\Desktop\water in Soil Plant System\Exercise\Exercise 8\completed\EC2_N-Balance\EC2_N-Balance\output\EC2_N-Balance_1_1_1_0_Organic.xno"

# --- labels for plotting ---
label1 = "Run 1_Mineral"    # Mineral Fertalization
label2 = "Run 2_Organic"    # Organic Fertalization

# --- column names (adjust to match your CSV headers) ---
col_time          = "Time and Date"
col_n_input       = "output.Balance.Soil N Balance.N Input[kg N/ha]"
col_n_mineralized = "output.Balance.Soil N Balance.Nmin mineralized [kg N/ha]"
col_n_immobilized = "output.Balance.Soil N Balance.Nmin immobilized [kg N/ha]"
col_n_uptake      = "output.Balance.Soil N Balance.Nmin Uptake [kg N/ha]"
col_gaseous       = "output.Balance.Soil N Balance.Gaseous N Emission [kg N/ha]"
col_leaching      = "output.Balance.Soil N Balance.N Leaching [kg N/ha]"

def compute_balance(filename):
    df = pd.read_csv(filename, sep=",")  # use sep=";" if needed
    df['Time and Date']=pd.to_datetime(df['Time and Date'])
    # cumulative sums
    df["N_input_cum"]       = df[col_n_input].cumsum()
    df["N_mineralized_cum"] = df[col_n_mineralized].cumsum()
    df["N_immobilized_cum"] = df[col_n_immobilized].cumsum()
    df["N_uptake_cum"]      = df[col_n_uptake].cumsum()
    df["N_gaseous_cum"]     = df[col_gaseous].cumsum()
    df["N_leaching_cum"]    = df[col_leaching].cumsum()
    
    # ΔN (detailed balance)
    df["DeltaN_detailed"] = (
        df["N_input_cum"] + df["N_mineralized_cum"]
        - (df["N_uptake_cum"] + df["N_immobilized_cum"]
           + df["N_gaseous_cum"] + df["N_leaching_cum"])
    )
    return df

# compute for both runs
df1 = compute_balance(file1)
df2 = compute_balance(file2)

# --- plot comparison ---
plt.figure(figsize=(10,6))
plt.plot(df1[col_time], df1["DeltaN_detailed"], label=label1, linewidth=2)
plt.plot(df2[col_time], df2["DeltaN_detailed"], label=label2, linewidth=2)
plt.xlabel("Time")
plt.ylabel("ΔN (kg N/ha)")
plt.title("Comparison of Nitrogen Balance (ΔN)")
plt.legend()

plt.tight_layout()
plt.show()

# --- print final values ---
print(f" ΔN for {label1}: {df1['DeltaN_detailed'].iloc[-1]:.2f} kg N/ha")
print(f" ΔN for {label2}: {df2['DeltaN_detailed'].iloc[-1]:.2f} kg N/ha")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- your two simulation files ---
file1 = r"C:\Users\moham\OneDrive\Desktop\water in Soil Plant System\Exercise\Exercise 8\completed\EC2_N-Balance\EC2_N-Balance\output\EC2_N-Balance_1_1_1_0_MIneral.xno"
file2 = r"C:\Users\moham\OneDrive\Desktop\water in Soil Plant System\Exercise\Exercise 8\completed\EC2_N-Balance\EC2_N-Balance\output\EC2_N-Balance_1_1_1_0_Organic.xno"

label1 = "Run 1_Mineral"
label2 = "Run 2_Organic"

col_time = "Time and Date"
col_fruit = "output.Plant.Biomass.Fruits [kg/ha]"
col_n_input = "output.Balance.Soil N Balance.N Input[kg N/ha]"
col_n_mineralized = "output.Balance.Soil N Balance.Nmin mineralized [kg N/ha]"
col_n_immobilized = "output.Balance.Soil N Balance.Nmin immobilized [kg N/ha]"
col_n_uptake = "output.Balance.Soil N Balance.Nmin Uptake [kg N/ha]"
col_gaseous = "output.Balance.Soil N Balance.Gaseous N Emission [kg N/ha]"
col_leaching = "output.Balance.Soil N Balance.N Leaching [kg N/ha]"

def compute_balance(filename):
    df = pd.read_csv(filename, sep=",")
    df[col_time] = pd.to_datetime(df[col_time])
    df["N_input_cum"] = df[col_n_input].cumsum()
    df["N_mineralized_cum"] = df[col_n_mineralized].cumsum()
    df["N_immobilized_cum"] = df[col_n_immobilized].cumsum()
    df["N_uptake_cum"] = df[col_n_uptake].cumsum()
    df["N_gaseous_cum"] = df[col_gaseous].cumsum()
    df["N_leaching_cum"] = df[col_leaching].cumsum()
    df["DeltaN_detailed"] = (
        df["N_input_cum"] + df["N_mineralized_cum"]
        - (df["N_uptake_cum"] + df["N_immobilized_cum"] 
           + df["N_gaseous_cum"] + df["N_leaching_cum"])
    )
    return df

df1 = compute_balance(file1)
df2 = compute_balance(file2)

#======Maximum Biomass and Time of them========== 
max_fruit1 = df1[col_fruit].max()
max_fruit2 = df2[col_fruit].max()
time_max1 = df1[col_time][df1[col_fruit].idxmax()]
time_max2 = df2[col_time][df2[col_fruit].idxmax()]

# --- plot ---
fig, ax1 = plt.subplots(figsize=(10,6))

# ΔN
ax1.plot(df1[col_time], df1["DeltaN_detailed"], label=f"{label1} ΔN", color="blue", linewidth=2)
ax1.plot(df2[col_time], df2["DeltaN_detailed"], label=f"{label2} ΔN", color="green", linewidth=2)
ax1.set_xlabel("Time")
ax1.set_ylabel("ΔN (kg N/ha)")
ax1.tick_params(axis='y', labelcolor='black')

# Fruit Biomass
ax2 = ax1.twinx()
ax2.plot(df1[col_time], df1[col_fruit], label=f"{label1} Fruit Biomass", color="red", linestyle="--", linewidth=2)
ax2.plot(df2[col_time], df2[col_fruit], label=f"{label2} Fruit Biomass", color="orange", linestyle="--", linewidth=2)
ax2.set_ylabel("Fruit Biomass (kg/ha)")
ax2.tick_params(axis='y', labelcolor='black')

# --- annotate max points ---
ax2.scatter(time_max1, max_fruit1, color="red", s=100, zorder=5)
ax2.scatter(time_max2, max_fruit2, color="orange", s=100, zorder=5)
ax2.annotate(f"{max_fruit1:.0f}", xy=(time_max1, max_fruit1), xytext=(10,10),
             textcoords="offset points", color="red")
ax2.annotate(f"{max_fruit2:.0f}", xy=(time_max2, max_fruit2), xytext=(10,-15),
             textcoords="offset points", color="orange")

# legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc="upper left")

plt.title("ΔN vs Fruit Biomass Comparison with Max Yield")
plt.tight_layout()
plt.show()
```
